```python
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import numpy as np

# The MNIST data (x_train, y_train, x_test, y_test) is already loaded and normalized (x_train/x_test values are between 0 and 1) in previous cells.
# We need to flatten the 28x28 images into a 1D array of 784 pixels for scikit-learn's MLPClassifier.
x_train_flattened = x_train.reshape(x_train.shape[0], -1)
x_test_flattened = x_test.reshape(x_test.shape[0], -1)

print("Shape of flattened x_train:", x_train_flattened.shape)
print("Shape of flattened x_test:", x_test_flattened.shape)

# Initialize the MLPClassifier
# We'll use a hidden layer size similar to the Keras model (128, 64)
# max_iter is increased for better convergence, alpha for L2 regularization
mlp_model = MLPClassifier(hidden_layer_sizes=(128, 64), max_iter=20, alpha=1e-4,
                          solver='adam', verbose=True, random_state=1,
                          learning_rate_init=.001)

# Train the model
print("\nTraining the MLPClassifier...")
mlp_model.fit(x_train_flattened, y_train)

print("\nTraining complete.")
```

```python
# Make predictions on the test set
y_pred_mlp = mlp_model.predict(x_test_flattened)

# Evaluate the model
accuracy_mlp = accuracy_score(y_test, y_pred_mlp)

print(f"Accuracy of MLPClassifier on MNIST test set: {accuracy_mlp:.4f}")

# Optionally, display a few misclassified images
print("\nDisplaying a few misclassified images:")

misclassified_indices = np.where(y_pred_mlp != y_test)[0]

plt.figure(figsize=(10, 5))
for i, index in enumerate(misclassified_indices[:5]): # Display first 5 misclassified
    plt.subplot(1, 5, i + 1)
    plt.imshow(x_test[index], cmap='gray')
    plt.title(f"True: {y_test[index]}\nPred: {y_pred_mlp[index]}")
    plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Plot the training loss curve from the MLPClassifier
plt.figure(figsize=(10, 6))
plt.plot(mlp_model.loss_curve_)
plt.title('MLPClassifier Training Loss over Iterations')
plt.xlabel('Iteration')
plt.ylabel('Loss')
plt.grid(True)
plt.show()

print("Note: Scikit-learn's MLPClassifier does not directly provide validation loss or accuracy curves during training in the same way Keras does. This plot shows the training loss only.")
```

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# Load the MNIST dataset
(X_train, y_train), (X_test, y_test) = keras.datasets.mnist.load_data()

# Preprocess the data
X_train = X_train.reshape(-1, 28 * 28).astype('float32') / 255.0 # Flatten and normalize
X_test = X_test.reshape(-1, 28 * 28).astype('float32') / 255.0
num_classes = 10 # MNIST has 10 classes (digits 0-9)

# One-hot encode the labels
y_train = keras.utils.to_categorical(y_train, num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes)

# Build the neural network model
model = keras.Sequential([
layers.Dense(256, activation='relu', input_shape=(28 * 28,)),
layers.Dropout(0.4), # Apply 40% dropout

layers.Dense(128, activation='relu'),
layers.Dropout(0.4), # Apply 40% dropout
layers.Dense(num_classes, activation='softmax') # Output layer for multi-class

])

# Compile the model
model.compile(
optimizer='adam',
loss='categorical_crossentropy',
metrics=['accuracy']
)

# Early stopping callback
early_stopping = keras.callbacks.EarlyStopping(
monitor='val_loss',
patience=5,
restore_best_weights=True
)

# Train the model with early stopping
history = model.fit(
X_train, y_train,
validation_split=0.2,
epochs=100, # Allow early stopping to control the number of epochs
batch_size=32,
callbacks=[early_stopping],
verbose=1
)

# Evaluate the model on the test set
test_loss, test_accuracy = model.evaluate(X_test, y_test, verbose=0)
print(f"Test Accuracy: {test_accuracy:.2f}")
```

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import matplotlib.pyplot as plt

(x_train, y_train), (x_test, y_test) =keras.datasets.mnist.load_data()

# Data Preprocessing
# Normalize pixel values (0-255 → 0-1)
x_train = x_train.astype("float32") / 255
x_test = x_test.astype("float32") / 255

# Reshape to (samples, height, width, channels)
x_train = np.expand_dims(x_train, -1)
x_test = np.expand_dims(x_test, -1)

# One-hot encode labels
num_classes = 10
y_train = keras.utils.to_categorical(y_train, num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes)

# Build CNN Model
model = keras.Sequential([

# First Convolution Layer
layers.Conv2D(32, kernel_size=(3,3), activation='relu',
input_shape=(28,28,1)),
layers.MaxPooling2D(pool_size=(2,2)),

# Second Convolution Layer
layers.Conv2D(64, kernel_size=(3,3), activation='relu'),
layers.MaxPooling2D(pool_size=(2,2)),

# Flatten layer
layers.Flatten(),

# Fully Connected Layer
layers.Dense(128, activation='relu'),

# Output Layer
layers.Dense(num_classes, activation='softmax')
])

# Compile Model
model.compile(
loss='categorical_crossentropy',
optimizer='adam',
metrics=['accuracy']
)

# Model Summary
model.summary()

# Train Model
history = model.fit(
x_train, y_train,
batch_size=128,
epochs=10,
validation_split=0.1
)

# Evaluate Model
test_loss, test_accuracy = model.evaluate(x_test, y_test)
print("Test Accuracy:", test_accuracy)
```
